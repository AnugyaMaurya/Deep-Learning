```python
# make your own image

import numpy as np
image007 = np.zeros((50, 50, 3), dtype=int)
```

```python
plt.imshow(image007)
```

```python
# Make this image white 
image007[:, :, :] = 255
plt.imshow(image007)
```

```python
image007[:, :, :] = 0 

# Red 
image007[:10, :10, 0] = 255

plt.imshow(image007)
```

```python
image007[:, :, :] = 0

# Red 
image007[:10, :10, 0] = 255
# Green 
image007[:10, 10:20, 1] = 255
# Blue 
image007[:10, 20:30, 2] = 255

# White = Red[0] + Green[1] + Blue[2] 
image007[20:30, 20:30, 0] = 255
image007[20:30, 20:30, 1] = 255
image007[20:30, 20:30, 2] = 255


# Yellow = Red[0] + Green[1] 
image007[20:30, 30:40, 0] = 255
image007[20:30, 30:40, 1] = 255

plt.imshow(image007)
```
