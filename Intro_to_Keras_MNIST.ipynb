```python
# Download the data - 
# --Training ----    ---- Testing--- 
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
 # This function will pick up images from train dataset and 
 # - plot the image 
 # - print the correct label

def plot_digits(image_num):
  image = x_train[image_num]
  label = y_train[image_num]
  plt.imshow(image, cmap="gray")
  print(f"Label: {label}")
```

```python
# Plot random images of similar digits 

def subplot_digits(digit=0, num=10):
  labels = np.where(y_train == digit)[0] # collect all images of that particular digit
  labels = np.random.choice(labels, num) # Randomly pick 10 images 
  
  plt.figure(figsize=(12, 10)) # Create a plt figure and define the size of graph
  for i in range(num):         # For loop for the num of images 
    plt.subplot(1, num, i+1)   # creates a sub plot 
    # plt.title(f"Digit: {digit}")
    plt.imshow(x_train[labels[i]], cmap="gray") # this plots the images
    
```

```python
subplot_digits(digit=6, num=15)
```
